```python
df['Col3'] = df[['Nombre', 'Ap paterno', 'Ap materno']].apply(lambda x: ''.join(x), axis=1)
df
```

```python
df['Col3_clean']=df['Col3'].map(lambda x: x.replace("\xa0", " ")).map(lambda x: x.replace(" ", ""))
```
