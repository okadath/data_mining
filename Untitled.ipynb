```python

import pandas as pd
 
df_base = pd.read_excel("Base_txns.xlsx",   engine='openpyxl')
df_base.head()
```

```python
 
df_Tiendas = pd.read_excel("Tiendas.xlsx",   engine='openpyxl')
df_Tiendas.head()
```

```python
from pandasql import sqldf
query="""SELECT * FROM df_Tiendas
CROSS JOIN df_base
WHERE df_Tiendas.ID_TIENDA = df_base.Tienda;"""

df_orders = sqldf(query)
df_orders
```

```python
df_orders['Col3'] = df_orders[['Nombre', 'Ap paterno', 'Ap materno']].apply(lambda x: ''.join(x), axis=1)

df_orders['Col3_clean']=df_orders['Col3'].map(lambda x: x.replace("\xa0", " ")).map(lambda x: x.replace(" ", ""))
df_orders
```

```python
query2=""" SELECT * FROM df_orders WHERE NOT df_orders.Col3_clean = 'MiltonRodríguezMuñoz'; """

df_orders = sqldf(query2)
df_orders["monto"]=df_orders["Monto en pesos"]
df_orders
```
